#### which one should u use? (pyplot vs matplotlib OO method)

* when plotting something quickly , its okay to use the pyplot method
* when plotting something more advanced, use OO

```python
over_50 = heart[heart['age']>50]
over_50.head()
```

```python
# Pyplot method
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c='target');
```

```python
# OO method
fig , ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c='target',
            ax= ax);
#ax.set_xlim([45,100])
```

```python
fig , ax = plt.subplots(figsize = (10,6))

scatter = ax.scatter(x=over_50['age'],
                    y=over_50['chol'],
                    c=over_50['target'])

#customize
ax.set(title="heart disease and Cholestrol Levels",
      xlabel= "Age",
      ylabel="Cholestrol")

ax.legend(*scatter.legend_elements(),title="Target")

ax.axhline(over_50['chol'].mean(),
          linestyle = "--");
```

```python
over_50
```

```python
# subplot of age , chol , thalach

fig , (ax1,ax2) = plt.subplots(2,1,figsize = (10,10),
                     sharex = True)

scatter = ax1.scatter(x=over_50['age'],
                    y=over_50['chol'],
                    c=over_50['target'],
                     cmap = "winter")

#customize
ax1.set(title="heart disease and Cholestrol Levels",
      ylabel="Cholestrol")

ax1.legend(*scatter.legend_elements(),title="Target")

ax1.axhline(y=over_50['chol'].mean(),
          linestyle = "--");

scatter = ax2.scatter(x=over_50['age'],
                    y=over_50['thalach'],
                    c=over_50['target'],
                     cmap ="summer")

#customize
ax2.set(title="heart disease and Max Heart Rate",
      xlabel= "Age",
      ylabel="Thalac")

ax2.legend(*scatter.legend_elements(),title="Target")

ax2.axhline( y=over_50['thalach'].mean(),
          linestyle = "--")

fig.suptitle("Heart Disease Analysis" , fontsize = 16 , fontweight="bold");
```
